```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import plotly.express as px
from matplotlib.pyplot import plot as plt
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('spotify_songs.csv', sep=',')
data = data.drop('liked', axis=1)
```

```python
X_songs = data.iloc[:, :].values
```

```python
scaler_songs = StandardScaler()
X_scaled = scaler_songs.fit_transform(X_songs)
```

```python
wcss = []

for i in range(1, 10):

    Kmeans_songs = KMeans(n_clusters=i, random_state=0)
    Kmeans_songs.fit(X_scaled)
    wcss.append(Kmeans_songs.inertia_)
```

```python
grafico = px.line(x = range(1, 10), y = wcss)
grafico.show()
```

```python
Kmeans_songs = KMeans(n_clusters=4, random_state=0)
rotulos = Kmeans_songs.fit_predict(X_scaled)
```

```python
data["cluster"] = rotulos

data.groupby("cluster").mean()
```

```python
pca = PCA(n_components=2)
X_scaled_pca = pca.fit_transform(X_scaled)
X_scaled_pca.shape
```

```python
grafico = px.scatter(x= X_scaled_pca[:, 0], y= X_scaled_pca[:, 1], color=rotulos,)
grafico
```
